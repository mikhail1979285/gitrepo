```python
import numpy as np
import pandas as pd
from pandas import read_csv, DataFrame
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('input/train.csv')
```

```python
data.describe()
```

```python
data.loc[data['Square']<15,:]
```

```python

data.loc[data['Square']<15,'Square'] = data['Square']*10
data.loc[data['Square']<15,'Square'] = data['Square']*4
data.loc[data['Square']<15,:]
```

```python
data['Price_m']=data['Price']/data['Square']
data.loc[data['Price_m']>11000,:]
```

### data = pd.get_dummies(data) data.head()




```python
data.loc[data['Rooms']<1,:]
data.loc[data['Rooms']<1,'Rooms'] = 1
```

```python
data.loc[data['HouseYear']>2019,:]
```

```python
data.loc[data['HouseYear'] ==20052011,'HouseYear'] = 2008
data.loc[data['HouseYear'] ==4968,'HouseYear'] = 2008


data.loc[data['HouseYear']>2019,:]
```

```python
data.loc[data['Rooms'] ==19,'Rooms'] = 4
data.loc[data['Rooms'] ==10,'Rooms'] = 6
data.loc[data['Id'] ==6358,'Rooms'] = 2
data.loc[data['Id'] ==7594,'Rooms'] = 2
data.loc[data['Id'] ==14003,'Rooms'] = 2
data.loc[data['Id'] ==5927,'Rooms'] = 2
data.loc[data['Rooms']>3,:]
```

```python

price_mean = data['Price_m'].mean()
price_mean_by_rooms = data.groupby(['Rooms'], as_index=False).agg({'Price_m':'mean'}).rename(columns={'Price_m':'mean_price_r'})
price_mean_by_distr_rooms = data.groupby(['DistrictId', 'Rooms'], as_index=False).agg({'Price_m':'mean'}).rename(columns={'Price_m':'mean_price_ID'})
price_social = data.groupby(['DistrictId','Social_1'], as_index=False).agg({'Price_m':'mean'}).rename(columns={'Price_m':'mean_price_s'})
price_helth = data.groupby(['DistrictId','Helthcare_2'], as_index=False).agg({'Price_m':'mean'}).rename(columns={'Price_m':'mean_price_h'})
price_helth
```

```python
def prepare_data(df, price_mean_by_distr_rooms, price_mean_by_rooms, price_mean,price_social,price_helth):
    df = pd.merge(df, price_social, on=['DistrictId', 'Social_1'], how='left')
    df = pd.merge(df, price_mean_by_distr_rooms, on=['DistrictId', 'Rooms'], how='left')
    df = pd.merge(df, price_helth, on=['DistrictId', 'Helthcare_2'], how='left')
    df = pd.merge(df, price_mean_by_rooms, on='Rooms', how='left')
  
  
    df['mean_price_ID'] = df['mean_price_ID'].fillna(df['mean_price_r'])
    df['mean_price_ID'] = df['mean_price_ID'].fillna(price_mean)
    df['mean_price_r'] = df['mean_price_r'].fillna(price_mean)
    df['mean_price_s'] = df['mean_price_s'].fillna(df['mean_price_r'])
    df['mean_price_h'] = df['mean_price_h'].fillna(df['mean_price_r'])
    return df

data = prepare_data(data, price_mean_by_distr_rooms, price_mean_by_rooms, price_mean,price_social,price_helth)
```
