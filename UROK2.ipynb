### 1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
a = [20,20,20]
c =0
b =np.array(a,dtype=int)
for i in b:
    c+=i**2
c = math.sqrt(c)   
print('Длина вектора',c)
    
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.



### код окружности

```python

x = np.linspace(-10, 10, 1000000)
R=1
y=np.sqrt(R**2-x**2)
y1=-np.sqrt(R**2-x**2)
plt.plot(x, y)
plt.plot(x, y1)
```

### код элипса

```python
x = np.linspace(-10, 10, 1000000)
a=8
b=2
y=np.sqrt((a**2*b**2-x**2*b**2)/a**2)
y1=-np.sqrt((a**2*b**2-x**2*b**2)/a**2)
plt.plot(x, y)
plt.plot(x, y1)
```

### код гиперболы

```python
x = np.linspace(-10, 10, 1000000)
a=4
b=1
y=np.sqrt((x**2/a**2-1)*b**2)
y1=-np.sqrt((x**2/a**2-1)*b**2)
plt.plot(x, y)
plt.plot(x, y1)
```

### 5. Задание (в программе) Нарисуйте трехмерный график двух параллельных плоскостей.



```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=Axes3D(fig)
x=np.arange(-5,5,0.5)
y=np.arange(-5,5,0.5)
x,y=meshgrid(x,y)
z=2*x+4*y
ax.plot_surface(x,y,z)
z1=2*x+4*y+15
ax.plot_wireframe(x,y,z1)
```
