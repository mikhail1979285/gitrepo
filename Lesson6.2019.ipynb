### Задание 2

```python
t = np.linspace(0,10,51)
t
```

```python
f =np.cos(t)
f
```

```python
plt.plot(t,f,color='g')
plt.title('График f(t)')
plt.show()
```

```python
%config InlineBackend.figure_format='svg'
```

```python
plt.plot(t,f,color='g')
plt.title('График f(t)')
plt.show
```

```python
plt.plot(t,f,color='g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show
```

```python
plt.axis([0.5,9.5,-2.5,2.5])
plt.plot(t,f,color='g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show
```
