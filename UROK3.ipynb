### 4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1




```python
from scipy.optimize import fsolve
def f(p):
    x,y=p
    return (y- x**2+1,np.exp(x)+x-x*y-1)
x1,y1= fsolve(f,(-2,2))
print(x1,y1)
x = np.linspace(-3, 3, 100)
plt.plot(x, x**2-1)
plt.plot(x, ((np.exp(x)-1)/x+1))
```
